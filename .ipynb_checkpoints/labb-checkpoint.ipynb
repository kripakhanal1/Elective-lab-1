# Is for loop is an iterator?


#### No, a for loop is NOT an iterator.
   #### - An iterator is an object that returns values one by one using next().
   #### - A for loop is a control structure used to repeat code.
   #### - The for loop uses an iterator internally, but it is not itself an iterator.

# iterator
#### - An iterator is an object that:

#### - Returns one value at a time

#### - Remembers its current position

#### - Stops when items are finished

# Generator 
#### - A generator is a special type of iterator that:

#### - Produces values one at a time

#### - Uses the keyword yield

#### - Saves memory (does not store all values at once)

# Decorator
## -> A decorator is a function that adds extra behavior to another function without changing its original code.

```python
def gen():
    yield 1
    yield 2
    yield 3

g= gen()

print(next(g))  
print(next(g))  
print(next(g))
```

```python
def my_decorator(func):
    def wrapper():
        print("my name is kripa")
        func()
        print("python is a programming language")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```

```python


def square_stream(n):
    for i in range(1, n + 1):
        yield i * i


n = int(input("Enter a number: "))

for square in square_stream(n):
    print(square)
```
