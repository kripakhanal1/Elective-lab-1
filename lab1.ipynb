```python
class Numbers:
    def __init__(self, limit):
        self.limit = limit          # store limit in object

    def display(self):
        i = 1
        while i <= self.limit:      # iteration using self.limit
            if i <= self.limit:     # ifâ€“else
                print(i)
            else:
                print("Done")
            i += 1

obj = Numbers(5)   # limit = 5
obj.display()

    
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```
